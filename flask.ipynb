```python
# Q1. What is Flask Framework? What are the advantages of Flask Framework?

# Answer->

# Flask is a lightweight web framework written in Python. It is designed to be simple, flexible, and easy to use. 
# Flask provides tools and features that make it easy to build web applications and APIs quickly and efficiently.

# Advantages of Flask Framework:

# Lightweight: Flask is a lightweight framework that does not require any additional libraries or tools to be installed. 
#              This makes it easy to use and deploy, and reduces the amount of code needed to build an application.

# Flexibility: Flask is a flexible framework that allows developers to customize and extend its functionality as needed. 
#              It provides a simple, minimalistic core that can be extended with third-party libraries and tools.

# Easy to use: Flask is designed to be easy to use and learn, with a simple and intuitive API. It provides a clear and concise documentation, 
#              and a large community of developers that can help with any questions or issues.

# Modular design: Flask is designed with a modular structure that allows developers to break their applications into smaller, more manageable components.
#                 This makes it easy to build and maintain large-scale applications and APIs.

# Built-in development server: Flask comes with a built-in development server that makes it easy to test and debug applications during development.

# RESTful request dispatching: Flask has built-in support for RESTful request dispatching, which makes it easy to build RESTful APIs that conform to best practices.

# Flask-WTF: Flask has built-in support for Flask-WTF, which is a simple and flexible form validation and rendering library that makes it easy to build and validate forms.

# Flask is a powerful and flexible framework that provides a simple and intuitive API, and a wide range of features and tools that make 
# it easy to build web applications and APIs quickly and efficiently.
```

```python
# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
# Jupyter Notebook.

# Answer->

from flask import Flask , render_template

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(debug = True,host="0.0.0.0")
```

```python
# Q3. What is App routing in Flask? Why do we use app routes?

# Answer->

# App routing in Flask is the process of defining URL patterns (routes) for an application or API.
# Routes are used to associate a URL with a specific function that handles the request and generates a response.

# In Flask, routes are defined using the @app.route() decorator. 
# The decorator is used to specify the URL pattern for a function that handles a specific route.

from flask import Flask , render_template , url_for

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'
if __name__=='__main__':
    app.run(debug = True,host="0.0.0.0")
```

```python
# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
# url_for() function.

Answer->

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page.'

@app.route('/DataScience_batch')
def DataScience_batch():
    return 'welcome to data science batch'


if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('DataScience_batch'))
        
# NOTE->        
# When the url_for() function is called with the name of a route (i.e., the name of the endpoint function), it returns the URL for that route. 
# In the example above, we have printed the URLs for each of the defined routes by calling the url_for() function with their respective endpoint
# function names.
```
